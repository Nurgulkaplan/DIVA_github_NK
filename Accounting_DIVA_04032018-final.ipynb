```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_path = "JBEI_ABF_DIVA_Cost_FY18 - final.csv"
df = pd.read_csv(csv_path)
df
```

```python
df = df.fillna('DNA Sequencing')
df
```

```python
df.columns
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
# df.count()
```

```python
# df.dtypes
```

```python
df['Cost']=df['Cost'].replace('[\$,]', '', regex=True).astype(float)
# df.dtypes
```

```python
# Total Cost per project ID
# Total Labor hours per project ID
Project_ID_df= df.groupby(["Project ID-Activity Code"]).sum()
Project_ID_df


```

**Total Cost per project ID

```python
Project_ID_df["Cost"].plot(kind="bar",figsize=(16,4), title="Total DIVA Service Cost",grid=True)
```

**Total Labor hours per project ID

```python
Project_ID_df["Hours"].plot(kind="bar",figsize=(16,4),title="Total DIVA Team Labor Hours",grid=True)
```
