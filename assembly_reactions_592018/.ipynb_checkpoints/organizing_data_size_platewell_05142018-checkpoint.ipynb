```python
import pandas as pd
import csv
```

```python
df = pd.read_csv("echo_test.csv")
df.head()
```

```python
df.columns
```

```python
new_df = pd.DataFrame(df["OUTPUT_WELL"])
new_df["REACTION_NUMBER"] = df["REACTION_NUMBER"]
new_df['EXPECTED_SIZE'] = df['EXPECTED_SIZE']

nimbus_df = new_df.sort_values(['EXPECTED_SIZE'])
nimbus_df = nimbus_df.reset_index(drop=True)
nimbus_df.head()
```

```python
well_letter = ["A","B","C","D","E","F","G","H"]
well_number = ["01","02","03","04","05","06","07","08","09","10","11","12" ]
plate_96 = []
for let in well_letter:
    for num in well_number:
          well_id = let+num     
          plate_96.append(well_id)
          
```

```python
plate_df = pd.DataFrame(plate_96)
plate_df["well_96"] = plate_96
plate_df.head()
```

```python
well_letter_384 = ["A","C","E","G","I","K","M","O"]

well_number_384 = ["01","03","05","07","09","11","13","15","17","19","21","23", ]
plate_384 = []
for let in well_letter_384:
    for num in well_number_384:
          well_id_384 = let+num     
          plate_384.append(well_id_384)
          


plate_df["well_384"]= plate_384
plate_df.head()
```

```python
result_df = pd.concat([nimbus_df,plate_df['well_96'],plate_df["well_384"]], axis=1)
result_df.head()
```

```python
result_df.columns
```

```python
result_df = result_df.dropna(axis=0, how='any')
result_df
```

```python
result_df.to_csv("new_output.csv")
```
