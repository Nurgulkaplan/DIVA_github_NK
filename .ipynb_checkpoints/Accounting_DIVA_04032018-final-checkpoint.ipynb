```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_path = "JBEI_ABF_DIVA_Cost_FY18 - final.csv"
df = pd.read_csv(csv_path)
# df.head()
```

```python
df = df.fillna('DNA Sequencing')
# df.head()
```

```python
# df.columns
```

```python
df = df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)
```

```python
# df.count()
```

```python
# df.dtypes
```

```python
df['Cost']=df['Cost'].replace('[\$,]', '', regex=True).astype(float)
# df.dtypes
```

```python
# Total Cost per project ID
# Total Labor hours per project ID
Project_ID_df= df.groupby(["Project ID-Activity Code","DIVA Service"]).sum()
Project_ID_df

```

**Total Reagent Cost per project ID

```python
Project_ID_df["Cost"].plot(kind="bar",figsize=(16,4), title="DIVA Service Reagent Cost",grid=True)
```

**Total Labor hours per project ID

```python
Project_ID_df["Hours"].plot(kind="bar",figsize=(16,4),title="DIVA Team Labor Hours",grid=True)
```

```python
df_DIVA_Service = df.groupby(["DIVA Service"]).sum()
df_DIVA_Service
```

```python
df_DIVA_Service["Cost"].plot(kind="pie",figsize=(6,6), title ="DIVA Service Reagent Cost",shadow= True,startangle=140)
```

```python
df_DIVA_Service["Number Rn/Construct/Sample"].plot(kind="pie",figsize=(6,6), title ="DIVA Service Usage",shadow= True,startangle=140)

```

```python
df_DIVA_Service["Hours"].plot(kind="pie",figsize=(6,6), title ="DIVA Service Labor Hours",shadow= True,startangle=140)

```
