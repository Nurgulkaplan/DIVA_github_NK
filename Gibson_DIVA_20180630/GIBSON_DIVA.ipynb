```python
import pandas as pd 
import csv
```

```python
df = pd.read_csv("j5_PCR_Gibson_20180628  - PCR_info.csv", )
df["ID_Number"] = df["ID_Number"].apply(lambda x: int(x) if x == x else "")
df["DNA_Length"] = df["DNA_Length"].apply(lambda x: int(x) if x == x else "")
df.head()
```

```python
Column_quarter = ['A1', 'C1', 'E1', 'G1', 'I1', 'K1', 'M1', 'O1','A3', 'C3', 'E3', 'G3', 'I3', 'K3', 'M3', 'O3', 'A5',
 'C5', 'E5', 'G5', 'I5', 'K5', 'M5', 'O5', 'A7', 'C7', 'E7', 'G7', 'I7', 'K7', 'M7', 'O7', 'A9', 'C9', 'E9', 'G9', 'I9',
 'K9', 'M9', 'O9']
```

```python
# pmols = (weight in ng) x 1,000 / (base pairs x 650 daltons)
df["pmols"] = (df.Concentration*1000)/(df.DNA_Length*650)
# Necessary volume(nanoliter) in order to have 0.01 pmols for each fragment
user_input = input("Enter desired pmols: ")
df["Volume"] = (float(user_input)/df.pmols)*1000
df["Volume"] = df["Volume"].apply(lambda x: int(x) if x == x else "")
df.head()
```

```python

df_parts =pd.read_csv("AssemblyPieces.csv")

cols = list(df_parts.columns)

for col in cols:
    if col.startswith('Assembly Piece'):
        df_parts[col] = df_parts[col].apply(lambda x: int(x) if x == x else "")
        

df_parts
# ADD filter function so it contains only gibson assembly method
```

```python
df_parts.iloc[0,4]
df_parts.shape
```

```python
design_list = df_parts["Name"]
len(design_list)

part_list_assembly = []
Destination_well_list = []

for row in range(df_parts.shape[0]):
    single_design_parts = []
    Destination_well = []
    for col in range(df_parts.shape[1]):
        cell_value = df_parts.iloc[row,col]
        dest = df_parts.iloc[row,0]
        if type(cell_value) == int:
            single_design_parts.append(cell_value)
            Destination_well.append(dest)
            
        else:
            pass
    part_list_assembly.append(single_design_parts)
    Destination_well_list.append(Destination_well)
    

            
            
print(part_list_assembly)
print(Destination_well_list)
```

```python
New_part_list = []
for i in range(len(part_list_assembly)):
    for y in range(len(part_list_assembly[i])):
        Ids = part_list_assembly[i][y]
        New_part_list.append(Ids)
print(New_part_list)
print(len(New_part_list))
```

```python
New_Dest_Well_List = []
for i in range(len(Destination_well_list)):
    for y in range(len(Destination_well_list[i])):
        Destinations = Destination_well_list[i][y]
        New_Dest_Well_List.append(Destinations)
print(New_Dest_Well_List)
len(New_Dest_Well_List)
```

```python
df.head()
```

```python
Source_well_list = []
Volume_list = []


for i in range(len(part_list_assembly)):
    for y in range(len(part_list_assembly[i])):
        ID  = part_list_assembly[i][y]
        for z in range(df.shape[0]):
            if ID == df.iloc[z,1]:
                source = df.iloc[z,0]
                vol = df.iloc[z,5]
            
                Source_well_list.append(source)
                Volume_list.append(vol)
            else:
                pass
            
print(Source_well_list)
print(Volume_list)

print(len(Source_well_list))
print(len(Volume_list))

```

```python
df_echo = pd.DataFrame(
    {'RN ID': New_part_list,
    'Source Well': Source_well_list,
     'Destination Well': New_Dest_Well_List,
     'Transfer Volume': Volume_list
    })
# df_echo
df_echo = df_echo[["RN ID","Source Well","Destination Well", "Transfer Volume"]]
df_echo.head()
```

```python
second_user_input = int(input("Enter total assembly reaction volume(microliter): "))
Gibson_mix_vol = second_user_input/2
Water_plus_DNA = second_user_input/2

Water_GibsonMix_vol_df = df_echo[["Destination Well","Transfer Volume"]]

Water_GibsonMix_vol_df = Water_GibsonMix_vol_df.groupby(["Destination Well"]).sum()

Water_GibsonMix_vol_df["Water Volume"] = (Water_plus_DNA*1000)-(Water_GibsonMix_vol_df["Transfer Volume"])
Water_GibsonMix_vol_df["Gibson Mix Volume"] = Gibson_mix_vol*1000
Water_GibsonMix_vol_df


```

```python
cols_inWG = list(Water_GibsonMix_vol_df.columns)

for col in cols_inWG:
    if col.endswith('Volume'):
        Water_GibsonMix_vol_df[col] = Water_GibsonMix_vol_df[col].apply(lambda x: int(x) if x == x else "")
        

Water_GibsonMix_vol_df.head()
```

```python
import collections

counter=collections.Counter(Rn_list)
print(counter)
# print(counter.values())
# print(counter.keys())
```

```python
new_df = pd.DataFrame(list(counter.keys()) )
new_df["counter"] = list(counter.values())
new_df["ID_Number"] = new_df[0]
new_df =new_df.drop(columns=[0])
new_df = new_df[["ID_Number","counter"]]
new_df.head()
```

```python
df_merged = pd.merge(df,new_df, on= "ID_Number")
df_merged.head(35)
```

```python
df_merged["Total_Volume"] = df_merged["Volume"]*df_merged["counter"]
df_merged.head()
```

```python
df_merged = df_merged.dropna(how='any')
df_merged.head()
```
