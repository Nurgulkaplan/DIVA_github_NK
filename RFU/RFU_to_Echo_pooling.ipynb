Needs plate number for downstream process. 

```python
import glob
list_of_rfu_file = glob.glob("*.xlsx")
list_of_rfu_file
```

```python
all_data = pd.DataFrame()
for f in list_of_rfu_file:
    df = pd.read_excel(f)
    all_data = all_data.append(df,ignore_index=True)
all_data.tail()
```

```python
max_rfu = all_data["End RFU"].max()
max_rfu
```

```python
all_data = all_data.loc[all_data["End RFU"] > 10 ]
```

```python
min_rfu = all_data["End RFU"].min()
min_rfu
```

# C1*V1 = C2*V2 

```python
all_data["Transfer Volume"] = (min_rfu*2000)/all_data["End RFU"]
all_data.tail()


```

```python
all_data["Transfer Volume"] = all_data["Transfer Volume"].apply(lambda x: 25 if x<25 else x)
all_data.tail()
```

```python
all_data["Transfer Volume"] = all_data["Transfer Volume"].round()
all_data.tail()
```

```python
type(all_data["Transfer Volume"][1])
```

```python
all_data["Transfer Volume"] = all_data["Transfer Volume"].astype(int)
all_data.tail()
```

```python
all_data.rename(columns={"Well": "Source Well"}, inplace=True)
all_data.tail()
```

```python
all_data["transfer_cumsum"] = all_data["Transfer Volume"].cumsum()
all_data.tail()
```

```python
all_data["bin_number"] = (all_data["transfer_cumsum"]//15000)+1
all_data.tail()
```

```python

# all_data['Destination Well'] = all_data[['transfer_cumsum']].apply(lambda x: 'A'.join(x), axis=1)
```

```python
all_data["Dest"] = "A"
```

```python
all_data.tail()
```

```python
all_data["bin_number"] = all_data["bin_number"].astype(str)
all_data.tail()
```

```python
all_data['Destination Well'] = all_data[["Dest","bin_number"]].apply(lambda x: ''.join(x), axis=1)
all_data.head()
```

# GENERATE ECHO POOLING INSTRUCTIONS 

```python
all_data['sort'] = all_data['Source Plate'].str.extract('(\d+)', expand=False).astype(int)
all_data.sort_values('sort',inplace=True, ascending=True)
all_data = all_data.drop('sort', axis=1)
all_data.tail()
```

```python
header = ['Source Plate', 'Source Well','Destination Well',
       'Transfer Volume', 
       ]

all_data.to_csv("Echo_pooling.csv", columns = header)
```
